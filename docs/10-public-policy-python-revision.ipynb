```python
from dataclasses import dataclass

unemployment_rates = [7.4, 7.1, 6.8, 6.9]
regions = ["Brussels", "Flanders", "Wallonia", "Belgium"]

region_to_rate = {region: rate for region, rate in zip(regions, unemployment_rates)}
above_seven = [region for region, rate in region_to_rate.items() if rate > 7.0]

@dataclass(frozen=True)
class ForecastScenario:
    name: str
    growth_rate: float
    inflation_rate: float

baseline = ForecastScenario(name="baseline", growth_rate=1.2, inflation_rate=2.1)

print(region_to_rate)
print("Regions above 7% unemployment:", above_seven)
print("Scenario:", baseline)
```

## 3. NumPy

`numpy` is the foundation for numerical computing in Python.

You use it for:
- arrays
- vectorized operations
- linear algebra
- numerical transformations
- fast calculations on many observations

In a public-policy context, `numpy` is useful when you need to:
- compute averages or variations over time
- normalize indicators
- work with matrices of sector-by-sector or region-by-region data
- prepare data for forecasting models


```python
import numpy as np

co2_emissions = np.array([102.0, 100.5, 97.8, 95.2, 93.9])
yearly_change = np.diff(co2_emissions)
average_emissions = co2_emissions.mean()
index_base_100 = (co2_emissions / co2_emissions[0]) * 100

print("Emissions:", co2_emissions)
print("Yearly change:", yearly_change)
print("Average emissions:", round(average_emissions, 2))
print("Index base 100:", np.round(index_base_100, 2))
```

## 4. Pandas

`pandas` is the classic library for tabular data analysis in Python.

You use it for:
- reading CSV/Excel files
- filtering rows
- grouping by categories
- computing indicators
- merging datasets
- working with time series

Typical policy-analysis examples:
- unemployment by region and year
- social benefits by household type
- energy use by sector
- demographic changes over time


```python
import pandas as pd

df = pd.DataFrame(
    {
        "year": [2021, 2021, 2022, 2022, 2023, 2023],
        "region": ["Brussels", "Flanders", "Brussels", "Flanders", "Brussels", "Flanders"],
        "unemployment_rate": [15.1, 4.8, 14.6, 4.5, 13.9, 4.3],
    }
)

summary = (
    df.groupby("region", as_index=False)["unemployment_rate"]
    .mean()
    .rename(columns={"unemployment_rate": "avg_unemployment_rate"})
)

brussels_only = df[df["region"] == "Brussels"]

print(df)
print()
print(summary)
print()
print(brussels_only)
```

## 5. Polars

`polars` is a modern dataframe library focused on performance and clear expressions.

Why interviewers may like it:
- it is fast
- it encourages explicit transformation pipelines
- it is well suited for large analytical datasets

In a policy-analysis role, `polars` is especially useful when datasets become large:
- survey microdata
- environmental measurements
- long multi-year socio-economic panels


```python
import polars as pl

pl_df = pl.DataFrame(
    {
        "year": [2021, 2021, 2022, 2022, 2023, 2023],
        "sector": ["industry", "services", "industry", "services", "industry", "services"],
        "energy_use": [120, 90, 116, 93, 111, 95],
    }
)

pl_summary = (
    pl_df.group_by("sector")
    .agg(pl.col("energy_use").mean().alias("avg_energy_use"))
    .sort("sector")
)

pl_filtered = pl_df.filter(pl.col("year") >= 2022)

print(pl_df)
print()
print(pl_summary)
print()
print(pl_filtered)
```

## 6. Statsmodels

`statsmodels` is the Python library most associated with **statistical modelling and inference**.

It is especially relevant for policy institutions because it is strong on:
- linear regression
- interpretable coefficients
- confidence intervals
- hypothesis testing
- time-series models

Typical use cases:
- estimate the relationship between inflation and consumption
- study the impact of education on employment
- forecast an indicator over time
- support a policy note with model interpretation


```python
import pandas as pd
import statsmodels.api as sm

reg_df = pd.DataFrame(
    {
        "public_investment": [10, 12, 13, 15, 18, 19, 21, 24],
        "employment_growth": [0.4, 0.6, 0.7, 1.0, 1.1, 1.4, 1.5, 1.8],
    }
)

X = sm.add_constant(reg_df[["public_investment"]])
y = reg_df["employment_growth"]
model = sm.OLS(y, X).fit()

print(model.summary())
```

## 7. Scikit-learn

`scikit-learn` is the standard machine-learning library in Python.

Use it for:
- preprocessing
- train/test split
- regression and classification models
- pipelines
- cross-validation
- evaluation metrics

For the target job, the important point is not only prediction accuracy, but also:
- reproducibility
- validation
- avoiding leakage
- explaining limits of the model


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

ml_df = pd.DataFrame(
    {
        "gdp_growth": [1.1, 1.5, 2.0, 1.8, 0.9, 1.2, 2.2, 2.5],
        "inflation": [1.9, 2.1, 2.4, 2.3, 3.1, 2.9, 2.0, 1.8],
        "employment_growth": [0.5, 0.7, 1.1, 0.9, 0.2, 0.4, 1.2, 1.4],
    }
)

X = ml_df[["gdp_growth", "inflation"]]
y = ml_df["employment_growth"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("MAE:", mean_absolute_error(y_test, predictions))
print("Predictions:", predictions)
```

## 8. Pytest

`pytest` is the testing library you use to protect your analysis and application logic.

In policy or forecasting work, testing matters because you want to avoid:
- broken transformations
- silent changes in formulas
- regressions in data pipelines
- wrong assumptions spreading into reports or dashboards

A good mental model:
- code computes indicators
- tests protect the meaning of those indicators


```python
def compute_dependency_ratio(active_population, retired_population):
    if active_population <= 0:
        raise ValueError("active_population must be positive")
    return retired_population / active_population


def test_compute_dependency_ratio():
    ratio = compute_dependency_ratio(100, 35)
    assert ratio == 0.35


def test_compute_dependency_ratio_rejects_invalid_input():
    try:
        compute_dependency_ratio(0, 35)
        assert False, "Expected ValueError"
    except ValueError:
        assert True


print("This cell shows pytest-style test functions.")
print("To run them for real, place them in a test_*.py file and launch pytest.")
```

## 9. PyQt

`PyQt` is used to build desktop graphical interfaces in Python.

It is less central for pure statistical analysis than pandas or statsmodels, but it can still matter when you need:
- a desktop internal tool
- a data-entry interface
- an analysis viewer for non-technical users

For interview revision, know the big picture:
- widgets build the interface
- signals and slots connect user actions to Python functions
- layouts organize the window


```python
pyqt_example = '''
from PyQt6.QtWidgets import QApplication, QLabel, QPushButton, QVBoxLayout, QWidget

app = QApplication([])
window = QWidget()
layout = QVBoxLayout()

label = QLabel("Policy dashboard")
button = QPushButton("Refresh indicators")
button.clicked.connect(lambda: label.setText("Indicators refreshed"))

layout.addWidget(label)
layout.addWidget(button)
window.setLayout(layout)
window.show()
app.exec()
'''

print(pyqt_example)
```

```python
django_example = '''
from django.db import models

class EconomicForecast(models.Model):
    year = models.IntegerField()
    scenario = models.CharField(max_length=100)
    gdp_growth = models.FloatField()
    inflation_rate = models.FloatField()

    def __str__(self):
        return f"{self.year} - {self.scenario}"
'''

print(django_example)
```
